## **패션 MNIST 데이터 셋**
- <font color=#ffff00>10 종류</font>의 패션 아이템으로 구성
- 매우 유명한 데이터셋
- <font color=#ffff00>텐서플로의 케라스 패키지</font> 임포트 후 데이터 로딩
- MNIST와 크기, 개수 동일하지만 숫자 대신 <font color=#ffff00>패션 아이템</font>

```python
from tensorflow import keras
(train_input, train_target), (test_input, test_target) =\
  keras.datasets.fashion_mnist.load_data()
```

> MNIST 데이터셋
- 0부터 9까지의 <font color=#ff0000>숫자</font>
- <font color=#ff0000>28 x 28</font> 크기 이미지
- 각 픽셀별로 <font color=#ff0000>~255 값</font>

## **데이터 시각화**
- 어떤 이미지인지 직접 보는 것이

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 20, figsize = (10, 10))
for i in range(20):
  axs[i].imshow(train_input[i], cmap='hot')
  axs[i].axis('off')
plt.show()
```

> - color map을 <font color=#ff0000>hot</font>으로 해서 칼라처럼 보이는 것
- <font color=#ff0000>gray_r<font color=#ff0000>로 하면 원래 흑백 이미지
