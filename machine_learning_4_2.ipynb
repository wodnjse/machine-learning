## **확률적**
- 내려올 때 가장 가파른 길을 찾는 방법: 훈련 세트를 사용함
- 전체 샘플을 사용하지 않고 딱 하나의 샘플을 훈련 세트에서 랜덤하게 골라 가장 가파른 길을 찾음!

- <font color= #F9D923>훈련 세트에서 랜덤하게 하나의 샘플</font>을 고르는 것: 확률적 경사 하강법

## **확률적 경사 하강법**
- **<font color= #F9D923>확률적</font>**: 무작위하게 or 랜덤하게
- **<font color= #F9D923>경사</font>**: 기울기
- **<font color= #F9D923>하강법</font>**: 내려가는 방법 → 경사를 따라 내려가는 방법

**[산에서 내려온다고 가정]**
1. 집으로 가려면 등산로 입구까지 내려가야 함
2. 가장 빠른 길을 선택하는 것이 좋음
3. 가장 빠른 길은 경사가 가장 가파른 길
<br>
<br>
---
<br>

1. **<font color= #F9D923>훈련 세트에서 랜덤하게 하나의 샘플</font>**을 선택하여 가파른 경사를 <font color= #F9D923>조금</font> 내려감
2. 그다음 훈련 세트에서 랜덤하게 또다른 샘플을 하나 선택하여 조금 내려감
3. 전체 샘플을 모두 사용할 때까지 <font color= #F9D923>계속 반복</font>

> 만약 모든 샘플을 다 사용했음에도 불구하고 산을 다 내려오지 못했으면?
 - **<font color= #F06E5E>다시 처음부터</font>** 시작 → 훈련 세트에 모든 샘플을 다시 채워 넣고 반복

> 다시 랜덤하게 하나의 샘플을 선택해 이어서 경사 내려감
 - **<font color= #F06E5E>만족할만한</font>** 위치에 도달할 때까지 계속 내려가는 것



- 훈련 세트를 한 번 모두 사용하는 과정: _**<font color= #F9D923>에포크(epoch)</font>**_
- 일반적으로 수십, 수백 번 이상 에포크 수행
- 랜덤으로 샘플을 선택하기 때문에 아주 조금씩 내려가야 함
- 확률적 경사 하강법은 생각보다 꽤 잘 동작
- <font color= #F06E5E>**신경망 알고리즘**</font>에서 반드시 사용!
- 신경망은 일반적으로 많은 데이터를 사용하기 때문에 한 번에 모든 데이터를 사용하기 어렵고, 모델이 매우 복잡함

## **럭키백의 확률

```python
import pandas as pd
# csv 파일을 데이터 프레임으로 저장(본인 경로 확인!)
fish = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/머신러닝/4_1 럭키백/fish.csv')
fish.head()
```

## **입력 데이터 생성**

> 입력 데이터 5개 → 7개의 생선 종류 중 하나를 예측


- Species 열: 타깃 / <font color= #F9D923>나머지 5개 열: 입력 데이터</font>
- 데이터프레임에서 원하는 열을 리스트로 나열: 5개 열만 선택
- 반환된 새로운 데이터프레임을 <font color= #F9D923>to_numpy</font>() 메서드로 넘파이 배열로 변경
- [ \] 2개 쓰는 이유: <font color= #F9D923>2차원 배열

```python
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

## **데이터 분류 - 훈련 세트와 테스트 세트**
- 2장에서 배웠듯이 머신러닝에서는 기본으로 데이터 세트 2개가 필요

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

## **SDG 제공하는 분류용 클래스**
-  SGDClassitier / sklearn.linear_mondel 패키지에서 임포트!



```python
from sklearn.linear_model import SGDClassifier
```

- 2개의 매개변수 지정
  - loss: 손실 함수의 종류 지정 (log=logistic loss function)
  - max_iter: 수행할 에포크 회수

```python
sc = SGDClassifier(loss = 'log', max_iter=10, random_state = 42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```
