## **확률적**
- 내려올 때 가장 가파른 길을 찾는 방법: 훈련 세트를 사용함
- 전체 샘플을 사용하지 않고 딱 하나의 샘플을 훈련 세트에서 랜덤하게 골라 가장 가파른 길을 찾음!

- <font color= #F9D923>훈련 세트에서 랜덤하게 하나의 샘플</font>을 고르는 것: 확률적 경사 하강법

## **확률적 경사 하강법**

1. **<font color= #F9D923>훈련 세트에서 랜덤하게 하나의 샘플</font>**을 선택하여 가파른 경사를 <font color= #F9D923>조금</font> 내려감
2. 그다음 훈련 세트에서 랜덤하게 또다른 샘플을 하나 선택하여 조금 내려감
3. 전체 샘플을 모두 사용할 때까지 <font color= #F9D923>계속 반복</font>

> 만약 모든 샘플을 다 사용했음에도 불구하고 산을 다 내려오지 못했으면?
 - **<font color= #F06E5E>다시 처음부터</font>** 시작 → 훈련 세트에 모든 샘플을 다시 채워 넣고 반복

> 다시 랜덤하게 하나의 샘플을 선택해 이어서 경사 내려감
 - **<font color= #F06E5E>만족할만한</font>** 위치에 도달할 때까지 계속 내려가는 것



- 훈련 세트를 한 번 모두 사용하는 과정: _**<font color= #F9D923>에포크(epoch)</font>**_
- 일반적으로 수십, 수백 번 이상 에포크 수행
- 랜덤으로 샘플을 선택하기 때문에 아주 조금씩 내려가야 함
- 확률적 경사 하강법은 생각보다 꽤 잘 동작
- <font color= #F06E5E>**신경망 알고리즘**</font>에서 반드시 사용!
- 신경망은 일반적으로 많은 데이터를 사용하기 때문에 한 번에 모든 데이터를 사용하기 어렵고, 모델이 매우 복잡함

## **그 외 경사 하강법**

**<font color= #F06E5E>[미니배치 경사 하강법]</font>** minibatch gradient descent
- 1개씩 말고  무작위로 <font color= #F9D923>여러 개</font>의 샘플을 사용해 경사 하강법을 수행하는 방식
- 실전에서 아주 많이 사용

**<font color= #F06E5E>[배치 경사 하강법]</font>** batch gradient descent
- <font color= #F9D923>전체</font> 샘플을 사용하는 방법
- 전체 데이터를 사용하기 때문에 가장 안정적인 방법
- 컴퓨터 자원을 많이 사용 or 한 번에 전체 데이터를 모두 못 읽을 수도
